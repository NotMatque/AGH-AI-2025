```python
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
test_img = cv.imread("data/agh-logo-rgb.jpg", 1)
test_img = cv.cvtColor(test_img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(test_img)
plt.title("AGH")
plt.show()
```
