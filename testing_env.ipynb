# Test środowiska
Jeżeli ten kod zostanie prawidłowo wykonany, środowisko jest ustawione poprawnie.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import torch

%matplotlib inline
```

### Test ``matplotlib`` oraz ``openCV``:
Po wykonaniu poniższego kodu powinno pojawić się logo _AGH_

```python
test_img = cv.imread("data/agh-logo-rgb.jpg", 1)
test_img = cv.cvtColor(test_img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(test_img)
plt.title("AGH")
plt.axis('off')
plt.show()
```
